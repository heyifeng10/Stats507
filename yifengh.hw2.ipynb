### Problem 2: Fun with Strings
1. Write a function called is_palindrome, which takes a string as its only argument, and returns a Boolean.

```python
def is_palindrome(letters):
    let = letters.lower().replace(" ","")
    let_rev = ""
    for i in range(len(let)-1,-1,-1):
        let_rev += let[i] 
    return let == let_rev
```

```python
print(is_palindrome("Was it a car or a cat I saw") == True)
print(is_palindrome("T A C O cat") == True)
```

2. Write a function is_abecedarian, which takes a single argument in the form of a string and returns True if the argument is abecedarian and False otherwise.

```python
def is_abecedarian(letters):
    let = letters.lower().replace(" ","")
    for i in range(len(let)-1):
        if let[i] > let[i + 1]:
            return False
    return True
```

```python
print(is_abecedarian("adder") == True)
print(is_abecedarian("cat") == False)
```

3. Write a function called double_vowels that takes a string as its only argument and returns that string with all the vowels duplicated.

```python
def double_vowels(letters):
    s ={"a","e","i","o","u","A","E","I","O","U"}
    let = ""
    for i in range(len(letters)):
        if letters[i] in s:
            let = let + letters[i]*2
        else:
            let += letters[i]
    return let    
```

```python
print(double_vowels('cat') == "caat")
print(double_vowels('audio') == "aauudiioo")
```

### Problem 3: Fun with Lists:
1. Write a function list_reverse that takes a list as an argument and returns that list, reversed.<br>

```python
def list_reverse(lst):
    if not isinstance(lst,list):
        raise TypeError("Please input list type")
    left = 0
    right = len(lst) - 1
    while left < right:
        lst[left], lst[right] = lst[right], lst[left]
        left += 1
        right -= 1
    return lst
```

```python
print(list_reverse([1,2,3]) == [3,2,1])
```

2. Write a function is_sorted that takes a sequence seq as its only argument and returns True if the sequence is sorted in ascending order and returns False otherwise.

```python
def is_sorted(seq):
    for i in range(len(seq)-1):
        if seq[i] > seq[i+1]:
            return False
    return True
```

```python
print(is_sorted([1,2,3,4]) == True)
print(is_sorted("abcd") == True)
```

3. Write a function called binary_search that takes two arguments, a list of integers t (which is guaranteed to be sorted in ascending order) and an integer elmt, and returns True if elmt appears in list t and False otherwise.

```python
def binary_search(t, elmt):
    if not t:
        return False 
    left = 0
    right = len(t) - 1
    while left <= right:
        middle = (left + right)//2
        if elmt > t[middle]:
            left = middle + 1
        elif elmt < t[middle]:
            right = middle - 1
        else:
            return True
    return False
```

```python
print(binary_search([2,4,6,8,9],9) == True)
print(binary_search([2,4,6,8,9],0) == False)
```

### Problem 4: More Fun with Strings
1. Write a function called char_hist that takes a string as its argument and returns adictionary whose keys are characters and values are the number of times each character appeared in the input.

```python
def char_hist(string):
    str_lower = string.lower()
    count = dict()
    for item in str_lower:
        if item in count:
            count[item] += 1
        else:
            count[item] = 1
    return count
```

```python
print(char_hist("gattaca") == {'a':3,'c':1,'g':1,'t':2})
```

2. Write a function called bigram_hist that takes a string as its argument and returns a dictionary whose keys are 2-tuples of characters and values are the number of times that pair of characters appeared in the string.

```python
def bigram_hist(string):
    str_lower = string.lower()
    count = dict()
    for i in range(len(str_lower)-1):
        if (str_lower[i],str_lower[i+1]) in count:
            count[(str_lower[i], str_lower[i+1])] += 1
        else:
            count[(str_lower[i], str_lower[i+1])] = 1
    return count
```

```python
print(bigram_hist("mississippi") == {('i','p'):1,('i','s'):2,('m','i'):1,('p','i'):1,('p','p'):1,('s','i'):2,('s','s'):2})
```

### Problem 5: Tuples as Vectors
1. Implement a function called vec_scalar_mult, which takes two arguments: a tuple of numbers (oats and/or integers) t and a number (oat or integer) s and returns a tuple of the same length as t, with its entries equal to the entries of t multiplied by s.<br>

```python
def check_tuple(t):
    if not isinstance(t, tuple):
        return False
    for num in t:
        if not isinstance(num,(int,float)):
            return False
    return True

def vec_scalar_mult(t,s):
    if isinstance(s,(int, float)) and check_tuple(t):
        lst = [s*num for num in t]
        return tuple(lst)
    else:
        raise TypeError("The input type is not correct")
```

```python
print(vec_scalar_mult((1,2),4) == (4,8))
print(vec_scalar_mult((1,2,3),-2) == (-2,-4,-6))
```

2. Implement a function called vec_inner_product which takes two \vectors" (i.e., tuples of foats and/or ints) as its inputs and outputs a foat corresponding to the inner product of these two vectors.

```python
def vec_inner_product(x,y):
    if check_tuple(x) and check_tuple(y):
        if len(x) == len(y):
            c = [m*n for m, n in zip(x,y)]
            return sum(c)
        else:
            raise ValueError("The two inputs' dimensions don't agree")
    else:
        raise TypeError("The input vector type is wrong")
```

```python
print(vec_inner_product((1,1),(1,2)) == 3)
print(vec_inner_product((1,1),(-1,2)) == 1)
```

###  Problem 6: More Fun with Vectors
1. Write a function is_valid_sparse_vector that takes one argument, and returns True if and only if the input is a valid sparse vector, and returns False otherwise.<br>

```python
def is_valid_sparse_vector(arg):
    if not isinstance(arg,dict):
        return False
    for key, val in arg.items():
        if not isinstance(key,int):
            return False
        if key < 0 or not isinstance(val, float):
            return False
    return True
```

```python
print(is_valid_sparse_vector({1:2.,6:1.,5:3.5,7:4.0},) == True)
print(is_valid_sparse_vector({-2:1.2,3:2.5,5:4.1},) == False)
```

2. Write a function sparse_inner_product that takes two \sparse vectors" as its inputs, and returns a foat that is the value of the inner product of the vectors that the inputs represent.

```python
def sparse_inner_product(s,v):
    if not is_valid_sparse_vector(s) or not is_valid_sparse_vector(v):
        raise TypeError("Input is not a valid sparse vector.")
    res = 0.0
    if len(s) < len(v):
        for key,val in s.items():
            if key in v:
                res = res + val*v[key]
    else:
        for key, val in v.items():
            if key in s:
                res = res + val*s[key]
    return res
```

```python
print(sparse_inner_product({1:1.,2:2.,3:3.},{1:1.,3:8.,7:5.1}) == 25)
```

1. Write a function called my_sum that takes any number of numeric (ints and floats) arguments, and returns the sum of its arguments.

```python
def my_sum(*args):
    sum = 0
    if not args:
        return sum
    for num in args:
        sum += num
    return sum
```

```python
print(my_sum(3.2,4.1,-3) == 4.3)
print(my_sum(3,4) == 7)
```

2. Write a function called reverse_tuple that takes a tuple as its only argument and returns a tuple that is the reverse of the input.

```python
def reverse_tuple(s):
    lst = []
    for num in range(len(s)-1,-1,-1):
        lst.append(s[num])
    return tuple(lst)
```

```python
print(reverse_tuple((3,4,1),) == (1,4,3))
print(reverse_tuple((4,1),) == (1,4))
```

3. Write a function called rotate_tuple that takes two arguments: a tuple and an integer, in that order.

```python
def rotate_tuple(t,n):
    if not isinstance(t,tuple):
         raise TypeError("The input type should be tuple.")
    while not isinstance(n, int):
        try:
            n = int(n)
            print("Your input is not integer type, but it has been rounded down to be {}.".format(n))
        except:
            raise TypeError("The input type shouold be numeric type.")
    lst = list(t)
    if n > len(lst):
        n = n % len(lst)
    lst2 = [None]*len(lst)
    for i in range(len(lst)):
        lst2[(i+n)%len(lst)] = lst[i]
    return tuple(lst2)
```

```python
# Rotate 9 is equal to rotate 2; Similarly, rotate -9 is equal to rotate -2.
print(rotate_tuple((1,2,3,4,5,6,7),9) == (6,7,1,2,3,4,5))
print(rotate_tuple((1,2,3,4,5,6,7),2) == (6,7,1,2,3,4,5))
print(rotate_tuple((1,2,3,4,5,6,7),-9) == (3,4,5,6,7,1,2))
print(rotate_tuple((1,2,3,4,5,6,7),-2) == (3,4,5,6,7,1,2))
```
